```python
# dynamic token
ratio = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8]
computation = [(1 + r + r**2 + r**3)/4 for r in ratio]
acc = [0.506, 0.568, 0.66, 0.667, 0.673, 0.674]
plt.plot(computation, acc, label='dynamic_token')

computation = [0.5, 0.5, 1]
acc = [0.53,0.50, 0.69]
plt.scatter(computation, acc, label='vanilla')

computation = [0.587901, 0.65766253, 0.71907597, 0.77644657, 0.8327019,  0.89221282, 0.92844256, 0.97379125]
acc = [0.61060815, 0.64278685, 0.66394411, 0.68018107, 0.69051368, 0.69169455, 0.69189136, 0.69149774]
plt.plot(computation, acc, label='early_exit')

plt.legend()
plt.show()
```
