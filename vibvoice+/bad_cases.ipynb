```python
import pickle
import matplotlib.pyplot as plt
import librosa
import numpy as np
with open('bad_cases.pkl', 'rb') as f:
    data = pickle.load(f)
for d in list(data.keys())[100:]:
    print(d, data[d])
    ori_wave = librosa.load(d.replace('V2S_tmp', 'V2S'), sr=16000)[0]
    ori_spec = librosa.feature.melspectrogram(y=ori_wave, sr=16000, )
    wave = librosa.load(d, sr=16000)[0]
    spec = librosa.feature.melspectrogram(y=wave, sr=16000)
    plt.subplot(221)
    plt.plot(wave)
    plt.subplot(222)
    plt.plot(ori_wave)
    plt.subplot(223)
    librosa.display.specshow(librosa.power_to_db(spec, ref=np.max))
    plt.subplot(224)
    librosa.display.specshow(librosa.power_to_db(ori_spec, ref=np.max))
    plt.show()
    break
```
